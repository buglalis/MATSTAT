```python
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
N     = 100 # Обьём выборки
theta = 25  # параметр
sampling = np.array([random.randint(1, theta) for _ in range(N)])
```

```python
A = np.zeros((N,5))
for i in range(N):
    A[i][0] = 2*np.average(sampling[:i+1])
    # A[i][1] = np.min(sampling[:i+1])
    A[i][1] = np.min(sampling[:i+1])*(i+2)
    A[i][2] = np.max(sampling[:i+1])*(i+2)/(i+1)
    A[i][3] = np.max(sampling[:i+1]) + np.max(sampling[:i+1])
    if i > 0:
        A[i][4] = sampling[0] + np.average(sampling[1:i+1])
    else:
        A[i][4] = sampling[0]
```

```python
n = np.arange(1, N+1, 1)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(n, np.abs(theta-A[:,0]), c='#FF2B2B',marker="+", mec='#FF2B2B',  ms=5, mew=2, label="1 оценка")
# ax.plot(n, np.abs(theta-A[:,1]), c='#ff8b0f', marker="+", mec="#ff8b0f", ms=10, mew=2, label="2 оценка(испр)")
ax.plot(n, np.abs(theta-A[:,2]), c='#287233', marker="+", mec='#287233', ms=5, mew=2, label="3 оценка(испр)")
ax.plot(n, np.abs(theta-A[:,3]), c='#F345B1', marker="+", mec='#F345B1', ms=5, mew=2, label="4 оценка")
ax.plot(n, np.abs(theta-A[:,4]), c='#000000', marker="+", mec='#000000', ms=5, mew=2, label="5 оценка")
ax.set_title('График разницы $|\\theta - \\tilde{\\theta}|$')
ax.set_ylabel('$|\\theta - \\tilde{\\theta}|$')
ax.set_xlabel('Количество элементов n')
ax.minorticks_on()
ax.grid()
ax.grid(which='major',
        color = 'k', 
        linewidth = 1)
ax.grid(which='minor', 
        color = 'k', 
        linestyle = '-.',
        linewidth = 0.5)
ax.legend()
# plt.show()
plt.savefig('./LaTex/images/ex_1/1.png')
```
